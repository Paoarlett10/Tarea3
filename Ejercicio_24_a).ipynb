```python
import matplotlib.pyplot as plt

# Datos
Datos_x = [1.02, 0.95, 0.87, 0.77, 0.67, 0.56, 0.44, 0.30, 0.16, 0.01]
Datos_y = [0.39, 0.32, 0.27, 0.22, 0.18, 0.15, 0.13, 0.12, 0.13, 0.15]

# Construir la matriz A y el vector z
A = []
z = []

# Lo hacemos con base a nuestra ecuación
#(a * Y**2) + (b * X * Y) + (c * X) + (d * Y) + e = X**2
for x, y in zip(Datos_x, Datos_y):
    A.append([y**2, x*y, x, y, 1])
    z.append(x**2)

# Resolver por mínimos cuadrados

import numpy as np
A = np.array(A)
z = np.array(z)

# Resolver el sistema normal: (A^T A)x = A^T z
AtA = A.T @ A
Atz = A.T @ z

# Invertir AtA
AtA_invertida = np.linalg.inv(AtA)

# Solución
parametros = AtA_invertida @ Atz

# Extraer parámetros
a, b, c, d, e = parametros

print("Parámetros orbitales estimados:")
print(f"a = {a:.6f}")
print(f"b = {b:.6f}")
print(f"c = {c:.6f}")
print(f"d = {d:.6f}")
print(f"e = {e:.6f}")
```

```python
# Graficar los puntos y la órbita ajustada
RangoX = np.linspace(-0.5, 1.2, 400)
RangoY = np.linspace(-0.1, 0.6, 400)
X, Y = np.meshgrid(RangoX, RangoY)

# Ecuación de la órbita
Z = a * Y**2 + b * X * Y + c * X + d * Y + e - X**2

#Grafica
plt.contour(X, Y, Z, levels=[0], colors='lightgreen', linewidths=2)
plt.scatter(Datos_x, Datos_y, color='green', label='Datos observados')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Órbita elíptica ajustada por mínimos cuadrados')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
